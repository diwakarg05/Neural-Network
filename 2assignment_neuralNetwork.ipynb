```python
import kagglehub
import pandas as pd
import os
import numpy as np

# Step 3: Download dataset (Titanic)
path = kagglehub.dataset_download("shubhamgupta012/titanic-dataset")
print("Path:", path)
```

```python
files = os.listdir(path)
csv_files = [f for f in files if f.endswith('.csv')]

print("CSV Files:", csv_files)

file_path = os.path.join(path, csv_files[0])
df = pd.read_csv(file_path)

print("\nOriginal Data:")
print(df.head())
print(df.columns)
```

```python
# Drop columns safely
cols_to_drop = ['Name', 'Ticket', 'Cabin', 'PassengerId']
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

# Handle missing values
if 'Age' in df.columns:
    df['Age'] = df['Age'].fillna(df['Age'].mean())

if 'Embarked' in df.columns:
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Convert categorical → numeric
df = pd.get_dummies(df)

# Split features & target
X = df.drop('Survived', axis=1)
y = df['Survived']

# Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nANN Preprocessed Shape:", X_scaled.shape)
```

```python
# Convert to numpy
X_array = np.array(X_scaled)

# Reshape to CNN format (samples, height, width, channel)
X_cnn = X_array.reshape(X_array.shape[0], X_array.shape[1], 1, 1)

print("\nCNN Input Shape:", X_cnn.shape)
```
