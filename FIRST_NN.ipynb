```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
# Load digits dataset (0-9)
digits = load_digits()
X = digits.data
y = digits.target.reshape(-1, 1)

```

```python
# One-hot encoding
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Network parameters
input_size = 64   # 8x8 images
hidden_size = 128
output_size = 10  # digits 0-9
```

```python
# Placeholders
X_ph = tf.placeholder(tf.float32, [None, input_size])
y_ph = tf.placeholder(tf.float32, [None, output_size])

# Weights and biases
W1 = tf.Variable(tf.random_normal([input_size, hidden_size]))
b1 = tf.Variable(tf.zeros([hidden_size]))

W2 = tf.Variable(tf.random_normal([hidden_size, output_size]))
b2 = tf.Variable(tf.zeros([output_size]))
```

```python
# Forward propagation
hidden_layer = tf.nn.relu(tf.matmul(X_ph, W1) + b1)
output_layer = tf.matmul(hidden_layer, W2) + b2

# Loss
loss = tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits_v2(logits=output_layer, labels=y_ph)
)
```

```python
# Optimizer
optimizer = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)

# Accuracy
correct_pred = tf.equal(tf.argmax(output_layer, 1), tf.argmax(y_ph, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

# Training
epochs = 50
batch_size = 32
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for epoch in range(epochs):
        for i in range(0, len(X_train), batch_size):
            batch_x = X_train[i:i+batch_size]
            batch_y = y_train[i:i+batch_size]

            sess.run(optimizer, feed_dict={X_ph: batch_x, y_ph: batch_y})

        if epoch % 10 == 0:
            loss_val, acc_val = sess.run(
                [loss, accuracy],
                feed_dict={X_ph: X_test, y_ph: y_test}
            )
            print(f"Epoch {epoch}, Loss: {loss_val:.4f}, Accuracy: {acc_val:.4f}")

    # Final accuracy calculation moved inside the session block
    final_acc = sess.run(accuracy, feed_dict={X_ph: X_test, y_ph: y_test})
    print("Final Test Accuracy:", final_acc)
```
